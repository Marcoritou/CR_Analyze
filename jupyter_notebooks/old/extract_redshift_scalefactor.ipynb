```python
import illustris_python as il
import numpy as np
import pandas as pd
```

```python
numsnap = np.arange(1,100,1)
basepath = '/home/tnguser/sims.TNG/TNG50-1/output'

snapnum = np.zeros(len(numsnap), dtype='float32')
redshifts = np.zeros(len(numsnap), dtype='float32')
scalefactor = np.zeros(len(numsnap), dtype='float32')
h = np.zeros(len(numsnap), dtype='float32')
omega_0 = np.zeros(len(numsnap), dtype='float32')
omega_lambda = np.zeros(len(numsnap), dtype='float32')


for i in range(len(numsnap)):
    header = il.groupcat.loadHeader(basepath, numsnap[i])
    snapnum[i] = numsnap[i]
    redshifts[i] = header['Redshift']
    scalefactor[i] = header['Time']
    h[i] = header['HubbleParam']
    omega_0[i] = header['Omega0']
    omega_lambda[i] = header['OmegaLambda']
```

```python
Snaps_df = pd.DataFrame()

Snaps_df['snapnum'] = snapnum
Snaps_df['redshift'] = redshifts
Snaps_df['scalefactor'] = scalefactor
Snaps_df['h'] = h[i]
Snaps_df['omega_0'] = omega_0
Snaps_df['omega_lambda'] = omega_lambda

Snaps_df.to_csv('snap_header.csv', index=False)
```
